```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.model_selection import train_test_split

# dropped null rows because saving data as csv makes null rows appear
df = pd.read_csv('dataset.csv') 
df.dropna(how="all", inplace=True)
df.shape
df
```

```python
from sklearn.preprocessing import StandardScaler

y = df["Class"]
X = df.drop(columns=["Class"])

scaler = StandardScaler()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state =42)
scaler.fit(X_train)
X_train
```

```python
# This allows us to normalize our data without data leaking
X_train.loc[:, scaler.feature_names_in_] = scaler.transform(X_train)
X_test.loc[:, scaler.feature_names_in_] = scaler.transform(X_test)
```

```python
X_train
```

## Finding Best Hyperparameters and Feature Importance

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

# Since the paper already said 5, this isn't needed but it will look nice to see that we did it anyways
# to validate the orginal model

pipe = Pipeline([
    ('classifier', KNeighborsClassifier()),
])

search_space = [
    {'classifier': [KNeighborsClassifier()],
    'classifier__n_neighbors': [1, 3, 5, 7, 9]
    }, 
]

# The output text is displayed via the verbose parameter, feel free to change if it's too overwhelming to look at
bestKNN = GridSearchCV(pipe, search_space, cv=5, verbose=3)
bestKNN.fit(X_train, y_train)
bestKNN.best_params_, bestKNN.best_score_
```

## Accuracy of Models w/o Feature Selection

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import ConfusionMatrixDisplay
def show_Accuracy(model, X_train, y_train, X_test, y_test):
    yhat_train = model.predict(X_train)
    yhat_test = model.predict(X_test)
    acc_train = accuracy_score(y_train, yhat_train)

    # calculate the test set error
    acc_test = accuracy_score(y_test, yhat_test) 

    print(f'training set accuracy (n={y_train.shape[0]}): {acc_train:4.3f}')
    print(f'test set accuracy (n={y_test.shape[0]}): {acc_test:4.3f}')
    print()
    print('classificiation report on test set performance')
    print(classification_report(y_test, yhat_test))
    print()

    with sns.axes_style('white'):
        ConfusionMatrixDisplay.from_predictions(y_test, yhat_test)
```

```python
show_Accuracy(bestKNN, X_train, y_train, X_test, y_test)
```

```python
show_Accuracy(naiveBayes, X_train, y_train, X_test, y_test)
```
