```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.model_selection import train_test_split

# dropped null rows because saving data as csv makes null rows appear
df = pd.read_csv('dataset.csv') 
df.dropna(how="all", inplace=True)
print(df.shape)
#df

X = df.drop(columns=["Class"])
y = df["Class"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state =42)
scaler.fit(X_train)
#X_train

# This allows us to normalize our data without data leaking
X_train.loc[:, scaler.feature_names_in_] = scaler.transform(X_train)
X_test.loc[:, scaler.feature_names_in_] = scaler.transform(X_test)
X_train
```

```python
from sklearn.feature_selection import SelectKBest, chi2

# Use chi2 as the scoring function
selector = SelectKBest(chi2, k=10)
X_new = selector.fit_transform(X, y)

# Now use X_new instead of X to train your classifiers
```
