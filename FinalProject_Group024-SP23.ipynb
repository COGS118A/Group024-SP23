# Abstract 
Accurate classification of mitochondrial morphology is imperative for comprehending cellular health and functionality. We present a robust methodology for classifying mitochondrial morphology and introduce novel classification models, augmenting the baseline approach. Using extracted microscopy data features under diverse drug conditions from [Deep Analysis of Mitochondria and Cell Health Using Machine Learning](https://www.nature.com/articles/s41598-018-34455-y), we identified three distinct mitochondrial classes: punctate, swollen, and networked. The baseline comprised a K-nearest neighbors (KNN) model (k=5) and a naive Bayes classifier, employing pre-engineered features outlined in the aforementioned publication. We evaluated these against support vector machines (SVM), decision trees, and multi-layer perceptron NN through classification accuracy rates, revealing their proficiency in classifying morphologies. Unsupervised PCA highlighted innate mitochondrial classes. After rigorous nested cross-validation, we identified that the Random Forest model is the most effective, differing from the original publication. Our study advanced the reliability and generalizability of mitochondria classification by comprehensively comparing established and novel machine learning techniques.

# Background
Mitochondria, dynamic organelles involved in ATP production and cellular regulation, play essential roles in eukaryotic cell fate, homeostasis, and disease<a name="Chan et al."></a>[<sup>[2]</sup>](#Channote). These organelles exhibit considerable variation in number, mass, and shape<a name="Ahmad et al."></a>[<sup>[1]</sup>](#Ahmadnote). There is limited understanding of mitochondrial behavior and the overall morphological patterns within cell populations under different drug conditions. Additionally, the potential patterns in morphology as biomarkers for drug response in human cells remains unclear <a name="Giedt et al."></a>[<sup>[3]</sup>](#Giedtnote).

To advance scientific knowledge and enable early disease diagnosis, the development of image-based analytical techniques for phenotyping mitochondrial morphology in various diseases is crucial<a name="Giedt, et al."></a>[<sup>[3]</sup>](#Giedtnote). Giedt and colleagues highlight the importance of such techniques in cancer research, including the analysis of cancer cell lines and patient-derived cells. Their investigations demonstrate that cancer cells of diverse origins exhibit distinct mitochondrial phenotypes, which remain relatively stable over time. These phenotypes correlate with cell metabolic measurements and can be altered by therapeutic interventions in drug-sensitive cancers<a name="Giedt et al."></a>[<sup>[3]</sup>](#Giedtnote). The findings shed light on the significance of mitochondrial dynamics in cancer biology and drug response. Based on these discoveries, the authors propose that image-based mitochondrial phenotyping can serve as biomarkers to assess cancer phenotypes and predict drug response.

Machine learning (ML) approaches offer valuable tools for improving discovery and decision-making processes in drug development, provided there is an abundance of high-quality data<a name="Vamathevan et al."></a>[<sup>[4]</sup>](#Vamathevannote). ML can optimize various stages of drug discovery, including target validation, identification of prognostic biomarkers, and analysis of digital pathology data in clinical trials<a name="Vamathevan et al."></a>[<sup>[4]</sup>](#Vamathevannote). Despite their accurate predictions and insights, challenges such as the interpretability and repeatability of ML-generated results limit their widespread application. Furthermore, generating comprehensive high-dimensional data remains an ongoing requirement. Addressing these challenges and validating ML approaches can facilitate data-driven decision making, ultimately accelerating the drug discovery and development process while reducing failure rates.

Due to the variable nature of mitochondrial phenotypes, efficient and accurate methods for classifying mitochondrial morphology are needed. In a study titled "Deep Analysis of Mitochondria and Cell Health Using Machine Learning," researchers aimed to classify distinct morphological classes of mitochondria, including punctate, swollen, and networked forms, each exhibiting unique characteristics<a name="Zahedi et al."></a>[<sup>[5]</sup>](#Zahedinote). 

Given that abnormal morphological dynamics can indicate disease, rapid and precise classification methods are crucial. The vast amounts of data acquired make human analysis impractical, necessitating the utilization of cutting-edge machine learning techniques. By harnessing the power of these methodologies, researchers can enhance diagnostic capabilities and expedite drug discovery processes.

Therefore, in this study, we present an extensive analysis of various ML techniques for classifying mitochondrial morphology. Our objective is to compare the performance of novel and established ML methodologies, including K-nearest neighbors (KNN) and Naive Bayes as baseline models, and evaluate the effectiveness of Support Vector Machines (SVM), Decision Trees, Multilayer Perceptrons Neural Networks (NN) in this task. By quantitatively assessing the performance of each model, we aim to provide insights into the most suitable ML approaches for mitochondria morphology classification.

# Proposed Solution

In order to accurately classify mitochondrial morphology types, this study proposes the utilization of novel classification models in comparison to the baseline approach. The proposed solution involves employing a sequential methodology that includes data normalization, feature selection techniques, rigorous testing, and model evaluation. The goal is to enhance the reliability and generalizability of the findings and contribute to the advancement of research in computational biology.

Data Normalization: The first step in the proposed solution is to normalize the microscopy data. Data normalization ensures that all features are on a similar scale, preventing any particular feature from dominating the classification process due to its larger magnitude. The standard technique for data normalization involves min-max scaling. The choice of normalization technique is based on the specific characteristics of the data and extracted features.

Feature Selection Techniques: Feature selection is crucial to improve the efficiency and effectiveness of the classification models. In this study, two-step feature selection technique is proposed: Principal Component Analysis (PCA) and Analysis of Variance (ANOVA) in combination with scikit-learn select k best.
- Principal Component Analysis (PCA): PCA is a dimensionality reduction technique that identifies the most informative features from the original dataset. It transforms the dataset into a new set of uncorrelated variables called principal components. By selecting a subset of the most significant principal components, the dimensionality of the dataset is reduced while retaining the maximum amount of information. PCA helps to eliminate redundant or irrelevant features, improving the model's performance.
- Analysis of Variance (ANOVA): ANOVA is a statistical technique used to identify significant differences between groups based on the variation in a particular feature. By calculating the F-statistic and p-value for each feature, ANOVA can determine which features are most discriminative in classifying the mitochondrial morphology types. Features with high F-statistic and low p-value are considered more relevant and informative for classification.

Baseline Methodology: The baseline methodology involves training two models: a K-nearest neighbors (KNN) model and a naive Bayes classifier. These models leverage pre-engineered features outlined in the publication mentioned in the abstract. The KNN model with k=5 and the naive Bayes classifier serve as benchmarks for comparison against the proposed models.
Proposed Models: To assess the efficacy of the proposed models, three additional classification models are suggested for comparison: support vector machines (SVM), decision trees, and Multi-layer Perceptron NN.
-  Support Vector Machines (SVM): SVM is a supervised learning algorithm that aims to find an optimal hyperplane that separates the data into different classes. SVM can handle high-dimensional data effectively and can capture complex relationships between features. By choosing an appropriate kernel function, such as linear, polynomial, or radial basis function (RBF), SVM can accurately classify the mitochondrial morphology types.
- Decision Trees: Decision trees are hierarchical structures that utilize a sequence of rules to classify data. They provide interpretable models and are capable of capturing nonlinear relationships between features. Decision trees can be regularized to prevent overfitting by setting parameters like the maximum depth, minimum samples per leaf, or minimum impurity decrease. The proposed decision tree model can effectively classify the mitochondrial morphology types based on the selected features.
- Multi-layer Perceptron Neural Network (NN): A Multi-layer Perceptron (MLP) NN is a type of artificial neural network that consists of multiple layers of interconnected neurons. MLP NNs are powerful models capable of learning complex patterns and nonlinear relationships in the data. By adjusting the number of hidden layers, the number of neurons per layer, and activation functions, an MLP NN can accurately classify the mitochondrial morphology types based on the selected features.

Evaluation and Comparison: The evaluation of the proposed models is based on the classification accuracy rate, which measures the percentage of correctly classified instances. This evaluation metric allows for a comprehensive assessment of each model's proficiency in accurately classifying the mitochondrial morphology types.

The proposed solution aims to contribute to the advancement of research in computational biology by providing a thorough comparison between established and novel machine learning methodologies. By employing rigorous techniques, such as data normalization, feature selection, and comprehensive evaluation, the study enhances the reliability and generalizability of the findings. The proposed models, SVM, decision trees, and MLP NN, offer alternative approaches that may outperform the baseline models in accurately classifying mitochondrial morphology types, thereby improving our understanding of cellular health and function.

# Nested Cross Validation Model Selection

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
# Define a list of models or algorithms to compare
models = [
    ('Random Forest', RandomForestClassifier()),
    ('KNN', KNeighborsClassifier()),
    ('SVM', SVC()),
    ('Perceptron', MLPClassifier(hidden_layer_sizes=(4, ))),
]

# Define a dictionary of hyperparameter grids for each model
param_grids = {
    'Random Forest': {
        'n_estimators': [70, 100, 150],
        'max_depth': [None, 5, 10],
        'min_samples_split': [2, 5, 10]
    },
    'KNN': {
        'n_neighbors': [1, 3, 5, 7, 9],
        'weights': ['uniform', 'distance'],
        'p': [1, 2]
    },
    'SVM': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf']
    },
    'Perceptron': {
        'alpha': [0.0001, 0.001, 0.01],
        'activation': ['logistic', 'relu']
    },
}

accuracies = pd.DataFrame({})
```

```python
def loopModel(model, grid, xx, yy): #model is the model, grid is the grid for that model
    NUM_TRIALS=3
    nested_scores = []
    fittedModels =[]
    av_accuracy =[]
    for i in range(NUM_TRIALS):
        inner_cv = KFold(n_splits=5, shuffle=True, random_state=i)
        outer_cv = KFold(n_splits=5, shuffle=True, random_state=i)
        
        clf = GridSearchCV(estimator=model, param_grid=grid, cv=inner_cv)
        clf.fit(xx, yy)
        
        nested_score = cross_val_score(clf, X=xx, y=yy, cv=outer_cv)
        for j in nested_score:
            nested_scores.append(j)
        av_accuracy.append(nested_score.mean())
        fittedModels.append(clf)
    return (fittedModels, nested_scores, av_accuracy)  
```

```python
#RF Nested Cross Validation
rfBest = loopModel(models[0][1], param_grids[models[0][0]], X_train, y_train)
accuracies["Random Forest"] = rfBest[1]
print(accuracies["Random Forest"])
rfBest = rfBest[0][rfBest[2].index(max(rfBest[2]))]
```

```python
#KNN Nested Cross Validation
knnBest = loopModel(models[1][1], param_grids[models[1][0]], X_train, y_train)
accuracies["KNN"] = knnBest[1]
print(accuracies["KNN"])
knnBest = knnBest[0][knnBest[2].index(max(knnBest[2]))]
```

```python
#SVM Nested Cross Validation
svmBest = loopModel(models[2][1], param_grids[models[2][0]], X_train, y_train)
accuracies["SVM"] = svmBest[1]
print(accuracies["SVM"])
svmBest = svmBest[0][svmBest[2].index(max(svmBest[2]))]
```

```python
#MLP Nested Cross Validation
mlpBest = loopModel(models[3][1], param_grids[models[3][0]], X_train, y_train)
accuracies["Perceptron"] = mlpBest[1]
print(accuracies["Perceptron"])
mlpBest = mlpBest[0][mlpBest[2].index(max(mlpBest[2]))]
```

```python
#Naive Bayes Nested Cross Validation
accuracies["Naive Bayes"] = loopModel(GaussianNB(), {}, X_train, y_train)[1]
nbBest = GaussianNB().fit(X_train, y_train)
accuracies.head(7)
```

# Best Model of RF
Since we determined that Random Forest is the best performing algorithm from the ones we compared, we're going to use the RF model that had the best accuracy for the rest of our project.

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import label_binarize

bestRF = rfBest.best_estimator_ # Gets the best hyperparameters from our NCV

y = label_binarize(y, classes=[1,2,3])

# Shuffle and split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

# OnevRest classifier
clf = OneVsRestClassifier(bestRF)
y_score = clf.fit(X_train, y_train).predict_proba(X_test)
```

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, roc_auc = dict(), dict(), dict()

n_classes = 3

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
from itertools import cycle
colors = cycle(['blue', 'red', 'green'])

fig, ax = plt.subplots(figsize=(10, 8))

for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=1.5,
             label='ROC curve of {0} (area = {1:0.2f})'
             ''.format(class_names[i], roc_auc[i]))
    
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic for mitochondrial multi-class data')
plt.legend(loc="lower right")
plt.show()
```
